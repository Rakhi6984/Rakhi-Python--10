# PARAMETERS

```python
def greet_student(student_name):
    print("Welcome" , student_name , "to the Python Day 3 From LW")
```

```python
greet_student("Tom")
```

```python
greet_student("Jack")
```

```python
greet_student("Pop")
```

```python
def show_order(item_name, quantity, price_per_item):
    total = quantity * price_per_item 
    print(item_name, "X", quantity, "=", total)
```

```python
show_order("Apple",10,100)
```

```python
show_order("Mango",1,70)
```

```python
def apply_discount(price, discount_percent=10):
    final_price = (price * discount_percent/100)
    print("Final Price After Discount Is:" , final_price)
```

```python
apply_discount(1000)
```

```python
apply_discount(1000, 25)
```

```python
apply_discount(2000, 50)
```

```python
apply_discount(500)
```
