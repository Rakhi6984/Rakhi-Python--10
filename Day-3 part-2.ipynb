```python
def add (a, b):
    return a + b

def sub (a, b):
    return a - b

def multiply (a, b):
    return a * b

def divide (a, b):
    return a / b
first_number = int(input("Enter your First Number:"))
second_number = int(input("Enter your Second Number"))
operator = input("Enter operation from *, +, -, /")
if operator == "+":
    add(first_number, second_number)
elif operator == "-":
    sub(first_number, second_number)
elif operator == "*":
    multiply(first_number, second_number)
elif operator == "/":
    divide(first_number, second_number)
else:
    print("Enter Valid Option from *, +, -, /")
```

```python
print(add(50,10))
```

```python
print("Addition is", add(50,10))
```

```python
def is_valid_age (age):
    if age >= 18:
        return True
    else:
        return False
```

```python
names = ["Tom", "Krish", "Harry"]
len(names)
```

```python
age = [20, 15,22]
```

```python
for i in range(len(names)):
    if is_valid_age(age[i]):
        print(names[i], "can Register")
    else:
    
        print(names[i], "Is too young")
```
