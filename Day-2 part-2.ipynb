## DISCOUNT OFFER

```python
purchase_amount=int (input("Enter Your Amount to Check the Discount"))
if purchase_amount >= 10000:
    print("You Receive the Discount of 20%")
elif purchase_amount >= 5000:
    print("You Receive the Discount of 10%")
elif purchase_amount >= 1000:
    print("You Receive the Discount of 5%")
else:
    print("Purchase min 1000 to receive the Discount")
```

# ATM CODE

```python
card_validity = True
correct_pin = 4321
account_balance = 100000

entered_pin = int(input("Enter Your PIN: "))
withdraw_amount = int(input("Enter Your Withdraw Amount: "))

if card_validity == True:
    if entered_pin == correct_pin:
        if withdraw_amount <= account_balance:
            print("Please Collect your Money")
        else:
            print("Withdraw Amount is greater than Account Balance")
    else:
        print("Enter Correct Pin:")
else:
    print("Invalid Card:")
```

## LOAN APPROVAL SYSTEM

```python
monthly_salary = int(input("Enter Your Monthly Salary"))
credit_score =  int(input("Enter Your Credit Score "))
acount_balance = int(input("Enter Your Account Balance"))
if monthly_salary >= 30000:
    if credit_score >= 700:
        if account_balance >= 20000:
            print("Loan Approved")
        else:
            print("Sorry You Have Very less Security Deposit")
    else:
            print("Sorry Your Credit Score is Low")
else:
        print("Sorry Loan Not Approved")
```
