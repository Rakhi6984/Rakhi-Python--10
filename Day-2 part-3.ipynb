## WHILE LOOP

```python
count = 1
```

```python
while count <= 5:
    print(count)
    count = count + 1
```

## PASSWORD ATTEMPTS

```python
current_password = "admin123"
attempt= 0
while attempt < 3:
    entered_password = input("Enter Your Password:")
    if entered_password == current_password:
        print("Login Success")
        break
    else:
        attempt = attempt + 1
        print("Wrong Password Entered Try Again ")

```

## COUNTER TIMER

```python
seconds = int(input("Enter Your Seconds:"))

while seconds >= 0:
    print(seconds)
    seconds -= 1

print("launch")
```
